```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/Dataset for Data Analytics.xlsx')
df.head()
```

```python
print("Rows and Columns:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
print(df['Date'].head())
```

```python
text_columns = [
    'Product',
    'ShippingAddress',
    'PaymentMethod',
    'OrderStatus',
    'CouponCode',
    'ReferralSource'
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

print("Text formatting cleaned!")
```

```python
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'])
df['TotalPrice'] = pd.to_numeric(df['TotalPrice'])
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

print(df[['UnitPrice', 'TotalPrice', 'Date']].head())
```
