#Project 1

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/Dataset for Data Analytics.xlsx')
df.head()
```

```python
print("Rows and Columns:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')

print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()

print("Duplicate Rows:", duplicates)
```

```python
print(df['Date'].head())
```

```python
text_columns = [
    'Product',
    'ShippingAddress',
    'PaymentMethod',
    'OrderStatus',
    'CouponCode',
    'ReferralSource'
]

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()

print("Text formatting cleaned!")
```

```python
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'])
df['TotalPrice'] = pd.to_numeric(df['TotalPrice'])
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

print(df[['UnitPrice', 'TotalPrice', 'Date']].head())
```

```python
df.describe()
```

```python
print("Quantity Mean:", df['Quantity'].mean())
print("Quantity Median:", df['Quantity'].median())
print("Quantity Count:", df['Quantity'].count())

print("\nUnit Price Mean:", df['UnitPrice'].mean())
print("Unit Price Median:", df['UnitPrice'].median())
print("Unit Price Count:", df['UnitPrice'].count())

print("\nTotal Price Mean:", df['TotalPrice'].mean())
print("Total Price Median:", df['TotalPrice'].median())
print("Total Price Count:", df['TotalPrice'].count())
```

```python
product_trend = df['Product'].value_counts()

print("Product Sales Trend:")
print(product_trend)

most_common = product_trend.idxmax()
most_common_count = product_trend.max()

print("\nMost Common Product:")
print(f"{most_common} ({most_common_count} orders)")

least_common = product_trend.idxmin()
least_common_count = product_trend.min()

print("\nLeast Common Product:")
print(f"{least_common} ({least_common_count} orders)")
```

```python
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df['TotalPrice'] < lower_bound) |
    (df['TotalPrice'] > upper_bound)
]

print("Number of Outliers:", len(outliers))
outliers.head()
```
